для того щоб працювати з Null  значеннями ми спочатку їх створимо

```python
import pandas as pd
import numpy as np

# Завантаження CSV файлу
df = pd.read_csv('/Bike Sharing.csv', encoding='latin1')

# Виведення оригінального DataFrame
print("Original DataFrame:")
print(df)

# Заміна значень 0 на NaN (null)
df.replace(0, np.nan, inplace=True)

# Виведення зміненого DataFrame
print("\nDataFrame after replacing 0 with NaN:")
print(df.head(20))
```

```python
new_df = df.dropna()
print(new_df.head(20).to_string())
```

Заповнення null значень у конкретному стовпці:
але з деякими маніпуляціями щоб отримати Null значення

```python
import pandas as pd
import numpy as np

# Завантаження CSV файлу
df = pd.read_csv('/Bike Sharing.csv', encoding='latin1')

# Заміна значень 0 на NaN (null)
df.replace(0, np.nan, inplace=True)

# Заповнення null значень у стовпці 'Working Day'
df['Working Day'] = df['Working Day'].fillna(777)

# Перевірка наявності null значень у стовпці 'Working Day' після fillna
print("\nNull values in 'Working Day' after fillna:")
print(df['Working Day'].isnull().sum())

# Виведення зміненого DataFrame
print("\nDataFrame after replacing null values in 'Working Day':")
print(df.head(20).to_string())

```
