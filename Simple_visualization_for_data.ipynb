```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Завантаження даних з CSV файлу
df = pd.read_csv('/content/organizations-100.csv')
```

```python
# Групування даних за роками (сума або середнє для кількості співробітників)
grouped_data = df.groupby('Founded')['Number of employees'].mean()

plt.figure(figsize=(14, 8))  # Збільшено розмір графіка
plt.plot(grouped_data.index, grouped_data.values, marker='o', linestyle='-', color='dodgerblue',
         markersize=10, linewidth=2, label='Employees')

# Додавання підписів до окремих точок
for year, employees in zip(grouped_data.index, grouped_data.values):
    plt.text(year, employees, f'{int(employees)}', fontsize=10, ha='right', va='bottom', color='darkblue')

# Налаштування підписів осей та заголовка
plt.xlabel('Founded Year', fontsize=14, labelpad=10)
plt.ylabel('Number of Employees (Avg)', fontsize=14, labelpad=10)
plt.title('Average Number of Employees by Founded Year', fontsize=16, pad=20)

# Додавання сітки для кращої читабельності
plt.grid(True, linestyle='--', alpha=0.7)

# Налаштування міток осей
plt.xticks(fontsize=12, rotation=45, ha='right')
plt.yticks(fontsize=12)

# Додавання легенди
plt.legend(fontsize=12, loc='upper left')

# Забезпечення компактності графіка
plt.tight_layout()

plt.show()
```

```python

# Вибір топ-10 галузей для відображення
top_industries = df['Industry'].value_counts().head(10)

plt.figure(figsize=(12, 8))  # Збільшено розмір графіка
bars = plt.barh(top_industries.index, top_industries.values, color='skyblue', edgecolor='black')

# Додавання підписів до кожного стовпчика
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.5, bar.get_y() + bar.get_height() / 2, f'{int(width)}', va='center', fontsize=10, color='black')

# Зміна підписів осей та назви графіка
plt.xlabel('Number of Companies', fontsize=14, labelpad=10)
plt.ylabel('Industry', fontsize=14, labelpad=10)
plt.title('Top 10 Industries by Number of Companies', fontsize=16, pad=20)

# Налаштування міток осей
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Додавання сітки для покращення читабельності
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Розширення меж осі X, щоб уникнути "вилазення" останнього показника
plt.xlim(0, top_industries.max() + 5)  # Додано 5 одиниць простору

# Забезпечення компактності графіка
plt.tight_layout()

plt.show()

```
