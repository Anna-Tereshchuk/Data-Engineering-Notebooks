**DataFrame**  у pandas - це двовимірна структура даних, схожа на двовимірний масив або таблицю з рядками та стовпцями.

Читання CSV файлів як DataFrames

Використовуйте функцію `pd.read_csv` для читання CSV файлу та перетворення його у DataFrame.
`pd.read_csv('path/to/your/file.csv'`) читає CSV файл і повертає DataFrame

````python
import pandas as pd


df = pd.read_csv('/content/sample_data/california_housing_train.csv')
print(df)
````

**Виведення всього DataFrame:**
Використовуйте метод `df.to_string()`, щоб вивести весь DataFrame.

````python
import pandas as pd


df = pd.read_csv('/content/sample_data/mnist_train_small.csv')
print(df.head(20).to_string())
````

**Читання JSON файлу**
Використовуйте функцію `pd.read_json` для читання JSON файлу та перетворення його у DataFrame.
